In OpenCV, We use the cv2.calcHist() function to compute the histogram of an image. We
can use this function to compute the histogram of a region of the image. To compute a
histogram of a region in the image first we define a mask. The white color in the mask is
for regions to examine in the original input image and the black color in the mask image is
for regions to ignore. Now we calculate the histogram passing this mask as a parameter to
the function.
Let’s quickly review the function:
cv2.calcHist(images, channels, mask, histSize, ranges)
• images: This is the image that we want to compute a histogram for.
• channels : A list of indexes, where we specify the index of the channel we want to
compute a histogram for. To compute a histogram of a grayscale image, the list
would be  [0]
• To compute a histogram for all three red, green, and blue channels, the channels list
would be [0, 1, 2]
• mask:  If a mask is provided, a histogram will be computed for masked pixels only. If
we do not have a mask or do not want to apply one, we can just provide a value
of None
• histSize: This is the number of bins we want to use when computing a histogram.
Again, this is a list, one for each channel we are computing a histogram for. The bin
sizes do not all have to be the same. Here is an example of 32 bins for each
channel:[32, 32, 32]
• ranges: The range of possible pixel values. Normally, this is [0, 256] (that is not a

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Function to calculate and display the histogram of an image
def calculate_and_show_hist(img, title, show=True):
    hist = cv2.calcHist([img], [0], None, [256], [0, 256])
    if show:
        plt.plot(hist, color='black')
        plt.title(f"Histogram: {title}")
        plt.xlabel('Pixel Intensity')
        plt.ylabel('Frequency')
        plt.show()
    return hist
```

```python
# Function to perform and display histogram equalization
def equalize_and_show(image_path, image_title):
    # Read the original image in grayscale
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    # Check if the image was read properly
    if img is None:
        raise FileNotFoundError(f"Error: The image at '{image_path}' could not be read. Please check the file path.")
    
    # Perform histogram equalization
    equalized_img = cv2.equalizeHist(img)
    
    # Calculate histograms for original and equalized images
    original_hist = calculate_and_show_hist(img, f"Original {image_title}", show=False)
    equalized_hist = calculate_and_show_hist(equalized_img, f"Equalized {image_title}", show=False)
    
    return img, equalized_img, original_hist, equalized_hist
```

```python
# Paths to the images
dark_image_path = "dark.png"
bright_image_path = "bright.png"
gray_image_path = "gray.png"

# Perform histogram equalization for each image type
dark_img, equalized_dark, dark_hist, eq_dark_hist = equalize_and_show(dark_image_path, "Dark Image")
bright_img, equalized_bright, bright_hist, eq_bright_hist = equalize_and_show(bright_image_path, "Bright Image")
gray_img, equalized_gray, gray_hist, eq_gray_hist = equalize_and_show(gray_image_path, "Gray Image")
```

```python
# Display all images and histograms for comparison
plt.figure(figsize=(18, 18))

# Dark image comparison
plt.subplot(3, 4, 1)
plt.title("Original Dark Image")
plt.imshow(dark_img, cmap='gray')
plt.axis("off")

plt.subplot(3, 4, 2)
plt.title("Histogram: Original Dark")
plt.plot(dark_hist, color='black')

plt.subplot(3, 4, 3)
plt.title("Equalized Dark Image")
plt.imshow(equalized_dark, cmap='gray')
plt.axis("off")

plt.subplot(3, 4, 4)
plt.title("Histogram: Equalized Dark")
plt.plot(eq_dark_hist, color='black')

# Bright image comparison
plt.subplot(3, 4, 5)
plt.title("Original Bright Image")
plt.imshow(bright_img, cmap='gray')
plt.axis("off")

plt.subplot(3, 4, 6)
plt.title("Histogram: Original Bright")
plt.plot(bright_hist, color='black')

plt.subplot(3, 4, 7)
plt.title("Equalized Bright Image")
plt.imshow(equalized_bright, cmap='gray')
plt.axis("off")

plt.subplot(3, 4, 8)
plt.title("Histogram: Equalized Bright")
plt.plot(eq_bright_hist, color='black')

# Gray image comparison
plt.subplot(3, 4, 9)
plt.title("Original Gray Image")
plt.imshow(gray_img, cmap='gray')
plt.axis("off")

plt.subplot(3, 4, 10)
plt.title("Histogram: Original Gray")
plt.plot(gray_hist, color='black')

plt.subplot(3, 4, 11)
plt.title("Equalized Gray Image")
plt.imshow(equalized_gray, cmap='gray')
plt.axis("off")

plt.subplot(3, 4, 12)
plt.title("Histogram: Equalized Gray")
plt.plot(eq_gray_hist, color='black')

plt.tight_layout()
plt.show()
```
